## All Images

```python
# load all images in a directory

all_image = list()

for theme in listdir('./uploads/'):
    for filename in listdir('./uploads/' + theme):
        # load image
        img_data = image.imread('./uploads/' + theme + '/' + filename)
        # store loaded image
        all_image.append(img_data)
        print('> loaded %s %s' % (filename, img_data.shape))
```

```python
Image.fromarray(all_image[0])
```

```python
len(all_image)
```

```python
# find maximum and minimum of image height and width across all the images

max_width = 0
min_width = 2000

max_height = 0
min_height = 2000

for img in all_image:
    if img.shape[0] > max_height:
        max_height = img.shape[0]
    
    if img.shape[0] < min_height:
        min_height = img.shape[0]
        
    if img.shape[1] > max_width:
        max_width = img.shape[1]
        
    if img.shape[1] < min_width:
        min_width = img.shape[1]
    
print(f'min_height = {min_height}')
print(f'max_height = {max_height}')
print(f'min_width = {min_width}')
print(f'max_width = {max_width}')
```
