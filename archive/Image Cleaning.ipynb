## Image Resizing

```python
# consolidating all the styles in a dictionary

style_dict = {'Contemporary': contemp, 
              'Eclectic': eclectic, 
              'Industrial': industrial, 
              'Minimalistic': minimalist, 
              'Modern': modern, 
              'Retro': retro, 
              'Scandinavian': scandi, 
              'Traditional': traditional, 
              'Transitional': transitional, 
              'Vintage': vintage}
```

```python
# find maximum and minimum of image height and width across all styles

max_width = 0
min_width = 2000

max_height = 0
min_height = 2000

for style in style_dict.values():
    for name, img in style.items():
        if img.shape[0] > max_height:
            max_height = img.shape[0]

        if img.shape[0] < min_height:
            min_height = img.shape[0]

        if img.shape[1] > max_width:
            max_width = img.shape[1]

        if img.shape[1] < min_width:
            min_width = img.shape[1]

print(f'min_height = {min_height}')
print(f'max_height = {max_height}')
print(f'min_width = {min_width}')
print(f'max_width = {max_width}')
```

Challenges:

- try to resize then pad, got issues because some images are longer width, some in height.
- try to pad before size, issue is small images get too much padding, resizing it ends up mainly all white or black

```python
# retain image aspect ratio

for style_name, style in style_dict.items():
    folder = './processed/' + style_name + '/'
    if not os.path.exists(folder):
        os.makedirs(folder)
    print(style_name)
    
    for img_name, img in style.items():
        print(f'processing {img_name}')
        pict = Image.fromarray(img)
        pict.thumbnail((256,256)) # resize with ratio intact

        frame = Image.new('RGB',
                     (256, 256),
                     (0, 0, 0))  # Black
        frame.paste(pict)  # Not centered, top-left corner
        
        
        frame.save(folder + img_name + '.jpg', quality=100)
        
    print('***********************************************************')
```

```python
# resize image

for style_name, style in style_dict.items():
    folder = './processed/' + style_name + '/'
    if not os.path.exists(folder):
        os.makedirs(folder)
    print(style_name)
    
    for img_name, img in style.items():
        print(f'processing {img_name}')
        pict = Image.fromarray(img)
        pict.thumbnail((256,256)) # resize with ratio intact

        frame = Image.new('RGB',
                     (256, 256),
                     (0, 0, 0))  # Black
        frame.paste(pict)  # Not centered, top-left corner
        
        
        frame.save(folder + img_name + '.jpg', quality=100)
        
    print('***********************************************************')
```
