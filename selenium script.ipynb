```python
import os
import sys
import time
from selenium import webdriver
from selenium.webdriver.support.wait import WebDriverWait
import urllib.request
import requests
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains


uploadsFolder = 'uploads'
if not os.path.exists(uploadsFolder):
    os.makedirs(uploadsFolder)
```

```python
def findImageWithinPostAndSave(driver,uploadStyleFolder):
    #buffer = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, '/html/body/div[8]/div/div/div/div/div/div[2]/div[1]/div[2]/div/div/ul/li/div/div/img')))
    time.sleep(3)

    try:
        next_button = driver.find_element_by_xpath('/html/body/div[8]/div/div/div/div/div/div[2]/div[1]/div[1]/div[2]')
        for i in range(8):
            ActionChains(driver).move_to_element(next_button).click(next_button).perform()
    except: 
        pass
        
    
    imgsrcs = driver.find_elements_by_xpath('/html/body/div[8]/div/div/div/div/div/div[2]/div[1]/div[2]/div/div/ul/li/div/div/img')
    

    for imgsrc in imgsrcs:
        try:
            src = imgsrc.get_attribute('src')
            print(src)
            img=requests.get(src)#fetch image
            imagename = os.path.join( uploadStyleFolder,src.split('/')[4]+".jpg")
            #print(imagename)
#             with open(imagename,'wb') as writer:#open for writing in binary mode
#                 writer.write(img.content)#write the image
            urllib.request.urlretrieve(src, imagename)
            #time.sleep(3)
        except Exception as e:
            #print("An exception occurred")
            print(e)
            continue

def iteratePropertyTypeAndSaveFiles(driver, uploadStyleFolder):
    propertyTypes = ['HDB','Condo','Landed']
    while len(propertyTypes)!=0:
        # click on property type
        driver.find_element_by_xpath('//*[@id="content"]/div/div[1]/div/div[1]/div[3]/div[1]/div/div/div[1]/div/div').click()

        all_proptype= driver.find_element_by_class_name("_1NV4foLTsZ5_Ibwd1NyAks").find_elements_by_css_selector('div')
        for a in all_proptype:
            print(a.text)
            if a.text in propertyTypes:
                print(propertyTypes)
                givenProperty = a.text
                propertyTypes.remove(givenProperty)
                print(propertyTypes)
                a.click()
                time.sleep(5)
                goThroughCarousel(driver,uploadStyleFolder)
                break


def goThroughCarousel(driver, uploadStyleFolder):
    time.sleep(3)
    # click first post
    carouselphotos = driver.find_element_by_xpath('//*[@id="content"]/div/div[1]/div/div[1]/div[4]/div/div/div[1]/div[1]').click()

    for i in range(1,101):
        print(i, housingStyle)
        findImageWithinPostAndSave(driver,uploadStyleFolder)
        if i % 12 == 0:
            time.sleep(10)
        #click next Project
        driver.find_element_by_xpath('/html/body/div[8]/div/div/div/div/div/div[1]/div[1]/div[2]/div').click()
        
    driver.find_element_by_xpath('/html/body/div[8]/div/div/div/div/div/div[1]/div[2]/div').click()

#######################################################################################

driver = webdriver.Chrome('chromedriver.exe')
# wait = WebDriverWait(driver, 20)

driver.maximize_window()
website_url = "https://qanvast.com/sg/interior-design-singapore"
driver.get(website_url)

housingStyles = ['Vintage']
#housingStyles = ['Scandinavian','Modern','Minimalistic','Contemporary','Industrial','Eclectic','Traditional','Transitional','Retro','Vintage']

#close opening popup box
driver.find_element_by_class_name("_3DwPqVqtU3w156y1my0QFf").click()


while len(housingStyles)!=0:

    # click on style filter
    driver.find_element_by_xpath('//*[@id="content"]/div/div[1]/div/div[1]/div[3]/div[1]/div/div/div[2]/div').click()
    # retrieve all styles available in filter
    all_styletype= driver.find_element_by_class_name("_1NV4foLTsZ5_Ibwd1NyAks").find_elements_by_css_selector('div')

    # looping through each style
    for a in all_styletype:
        print(a.text) # print current style
        if a.text in housingStyles:
            housingStyle = a.text
            print(housingStyles)
            housingStyles.remove(housingStyle)
            print(housingStyles)
            a.click() # click on the current style to filter
            time.sleep(5) # sleep to let browser load

            uploadStyleFolder = os.path.join(uploadsFolder, housingStyle) 

            if not os.path.exists(uploadStyleFolder):
                os.makedirs(uploadStyleFolder)
            iteratePropertyTypeAndSaveFiles(driver,uploadStyleFolder)
            break
           


driver.close()
```
