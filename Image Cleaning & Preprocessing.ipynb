```python
style_list = [contemp, eclectic, industrial, minimalist, modern, retro, scandi, traditional, transitional, vintage]
```

```python
# find maximum and minimum of image height and width across all styles

max_width = 0
min_width = 2000

max_height = 0
min_height = 2000

for style in style_list:
    for img in style:
        if img.shape[0] > max_height:
            max_height = img.shape[0]

        if img.shape[0] < min_height:
            min_height = img.shape[0]

        if img.shape[1] > max_width:
            max_width = img.shape[1]

        if img.shape[1] < min_width:
            min_width = img.shape[1]

print(f'min_height = {min_height}')
print(f'max_height = {max_height}')
print(f'min_width = {min_width}')
print(f'max_width = {max_width}')
```
